```python
import matplotlib.pyplot as plt

# Data from the table (represented as lists for easy plotting)
providers = ['AWS', 'Google Cloud', 'Microsoft Azure', 'Open-Source CCF/Boavizta']
direct_indirect = [1, 1, 1, 1]  # All providers include Direct & Indirect

# Scope 3 data (using 1 to indicate inclusion, 0 for not applicable)
scope3 = [0, 1, 1, 1]
google_scope3_specific = [0, 1, 0, 0] # 1 if specific scope 3 data mentioned, otherwise 0
msft_scope3_specific = [0, 0, 1, 0]
opensource_scope3_specific = [0, 0, 0, 1]

# Create the stacked bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

# Plot Direct & Indirect emissions (base of the bars)
plt.bar(providers, direct_indirect, label='Direct & Indirect (Scope 1 & 2)', color='skyblue')

# Plot Scope 3 emissions on top
plt.bar(providers, scope3, bottom=direct_indirect, label='Scope 3 Upstream & Downstream / Embedded', color='lightcoral')

# Add specific Scope 3 segments (if applicable)
google_bottom = [sum(x) for x in zip(direct_indirect, scope3)]
plt.bar(providers, google_scope3_specific, bottom=google_bottom, label='Employee Commutes & DC HW Emissions (Google)', color='coral')

msft_bottom = [sum(x) for x in zip(direct_indirect, scope3)]
plt.bar(providers, msft_scope3_specific, bottom=msft_bottom, label='HW Lifecycle Emissions (Microsoft)', color='salmon')

opensource_bottom = [sum(x) for x in zip(direct_indirect, scope3)]
plt.bar(providers, opensource_scope3_specific, bottom=opensource_bottom, label='Server Embedded Emissions (Open Source)', color='darksalmon')


# Customize the chart
plt.xlabel('Cloud Provider')
plt.ylabel('Emissions Scope')
plt.title('Carbon Emissions Breakdown by Provider')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend(loc='upper right', fontsize='small')  # Adjust legend position and size
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Show the chart
plt.show()

# If you want to save the figure:
# plt.savefig("carbon_emissions_breakdown.png", dpi=300)
```
