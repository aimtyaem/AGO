```python

import yaml

class SustainabilityTracker:
    def __init__(self):
        self.scope2_emissions = 0.0
        self.energy_consumption_history = []
        self.recommendation = None
        self.roi_estimate = None
        self.annual_savings = None

    def select_goal(self, goal):
        print(f"Goal Selected: {goal}")

    def ask_questions(self):
        print("Answering manufacturing-specific questions...")

    def chatbot_calculate_emissions(self, kwh_usage=None, bill_path=None):
        if kwh_usage:
            print(f"Calculating emissions from {kwh_usage} kWh usage...")
            self.scope2_emissions += kwh_usage * 0.5  # Example calculation factor
        elif bill_path:
            print(f"Processing energy bill from {bill_path}...")
            # Demo assumption: extracted 1200 kWh from bill
            self.scope2_emissions += 1200 * 0.5
        else:
            print("Please provide kWh usage or upload an energy bill.")

    def detect_anomaly(self, current_kwh):
        if self.energy_consumption_history:
            average = sum(self.energy_consumption_history) / len(self.energy_consumption_history)
            if current_kwh > average * 1.25:
                print("WARNING: Significant energy consumption spike detected!")
        self.energy_consumption_history.append(current_kwh)

    def recommend_solution(self):
        self.recommendation = "Switch to LED Lighting"
        self.roi_estimate = "2 years"
        self.annual_savings = "$10,000"
        print(f"Recommendation: {self.recommendation}")
        print(f"Estimated ROI: {self.roi_estimate}")
        print(f"Estimated Annual Savings: {self.annual_savings}")

    def save_to_yaml(self, filename):
        data = {
            'scope2_emissions': self.scope2_emissions,
            'energy_consumption_history': self.energy_consumption_history,
            'recommendations': {
                'action': self.recommendation,
                'roi': self.roi_estimate,
                'annual_savings': self.annual_savings
            } if self.recommendation else None
        }
        with open(filename, 'w') as file:
            yaml.dump(data, file, default_flow_style=False)

# Example Usage
tracker = SustainabilityTracker()
tracker.select_goal("Reduce Scope 2 Emissions")
tracker.ask_questions()
tracker.chatbot_calculate_emissions(kwh_usage=1000)  # User provides kWh
tracker.detect_anomaly(1250)  # Current month shows a spike
tracker.recommend_solution()
tracker.save_to_yaml('sustainability_data.yaml')
```
